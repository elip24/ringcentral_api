```python
from configparser import ConfigParser
import os
from configparser import ConfigParser
import ast
from pathlib import Path

import polars

main_path=os.getcwd()
ROOT_PATH = [f.path for f in os.scandir(main_path) if f.is_dir()][2]
config=ConfigParser(interpolation=None)
config_file_path=os.path.join(ROOT_PATH,'config.ini')
config.read(config_file_path)
api_creds=config["RingCentral"]
```

```python
import json
import sys
import configparser
from ringcentral import SDK

# Load credentials from config.ini
config = configparser.ConfigParser()
config.read("config.ini")

# Instantiate SDK
rcsdk = SDK(
    api_creds['clientId'],
    api_creds['clientSecret'],
    api_creds['server']
)
platform = rcsdk.platform()
all_records=[]
# Read call log
def read_user_calllog():
    try:
        queryParams = {
            'dateFrom': "2025-07-09T00:00:00.000Z",
            'dateTo': "2025-07-09T23:59:59.009Z",
            'view': "Detailed"
        }
        endpoint = "/restapi/v1.0/account/~/call-log"
        resp = platform.get(endpoint, queryParams)
        jsonObj = resp.json_dict()
        for record in jsonObj.get('records', []):
            all_records.append(record)
        return all_records
    except Exception as e:
        print("Unable to read user call log. " + str(e))
        raise



# Login and run
def login():
    try:
        platform.login(jwt=api_creds['jwt'])
        return read_user_calllog()
    except Exception as e:
        sys.exit("Unable to authenticate to platform. " + str(e))

records=login()
```
